2. Implementasikan transformasi contrast pada Google Colaboratory menggunakan
formula yang terdapat pada bagian Ulasan Teori untuk kontras, sehingga
menghasilkan keluaran seperti berikut:

```python
print(' Mengubah kontras dan tingkat kecerahan citra ')
print('----------------------------------------------')

# Mengambil input dari pengguna untuk tingkat kecerahan dan kontras
try:
  brightness = int(input('Masukan tingkat kecerahan [-255 - 255]:'))
  contrast = float(input('Masukkan kontras [1.0 - 3.0]:'))
except ValueError:
  print('Error, not a number')

# Membaca citra asli
original = cv.imread('/content/drive/MyDrive/female.JPG')

# Membuat citra kontras yang kosong dengan tipe data yang sama seperti citra asli
contrast_image = np.zeros(original.shape, original.dtype)

# Mengiterasi melalui setiap piksel pada citra
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      # Menghitung nilai piksel yang baru dengan mengaplikasikan kontras dan tingkat kecerahan
      contrast_image[y,x,c] = np.clip(contrast * original[y,x,c] + brightness, 0, 255)

# Menggabungkan citra asli dengan citra kontras
final_frame = cv.hconcat((original, contrast_image))

# Menampilkan hasilnya menggunakan cv2_imshow
cv2_imshow(final_frame)
```

3. Implementasikan transformasi logarithmic brightness pada Google Colaboratory
menggunakan formula yang terdapat pada bagian Ulasan Teori untuk transformasi
log, sehingga menghasilkan keluaran seperti berikut:

```python
print('Mengubah tingkat kecerahan citra dengan Transformasi Log')
print('----------------------------------------------')

try:
  brightness = int(input('Masukkan nilai kecerahan:'))
except ValueError:
  print('Error, not a number')

# Membaca citra asli
original = cv.imread('/content/drive/MyDrive/female.JPG')

# Membuat citra baru untuk menyimpan citra dengan tingkat kecerahan yang diubah
brightness_image = np.zeros(original.shape, original.dtype)

# Melakukan transformasi log pada setiap pixel dalam citra asli
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      # Transformasi log diaplikasikan pada nilai piksel
      brightness_image[y, x, c] = np.clip(brightness * (np.log(1 + original[y, x, c])), 0, 255)

# Menggabungkan citra asli dan citra dengan tingkat kecerahan yang diubah secara horizontal
final_frame = cv.hconcat((original, brightness_image))

# Menampilkan hasilnya
cv2_imshow(final_frame)
```

4. Implementasikan transformasi grayscale menggunakan metode averaging, lightness,
dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada
bagian Ulasan Teori, sehingga menghasilkan keluaran seperti berikut:

```python
img = cv.imread('/content/drive/MyDrive/lion.JPG')
img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB) # konversi channel BGR -> RGB
plt.imshow(img_rgb)
```

```python
 #a Averaging
img_avg = cv.imread('/content/drive/MyDrive/lion.JPG')
# Mengambil rata-rata dari semua saluran warna untuk menghasilkan gambar grayscale
img_avg[:] = img_avg.mean(axis=-1, keepdims=1)

# Menampilkan gambar asli dan hasil averaging
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img_rgb)
ax[1].imshow(img_avg)
```

```python
#b Lightness

img_light = cv.imread('/content/drive/MyDrive/lion.JPG')
# Menghitung rata-rata dari saluran warna merah dan saluran warna hijau dari gambar
# dengan mengambil nilai maksimum dalam setiap saluran warna dan kemudian
# menggabungkannya dengan nilai minimum dalam setiap saluran warna.
img_light[:] = np.max(img_light, axis=-1, keepdims=100) / 2 + np.min(img_light, axis=-2, keepdims=100) / 3

# Menampilkan gambar asli dan hasil pemrosesan
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img_rgb)
ax[1].imshow(img_light)
```

```python
#c Luminance
img_lum = cv.imread('/content/drive/MyDrive/lion.JPG')
# Mengambil saluran warna merah, hijau, dan biru dari citra img_lum
R, G, B = img_lum[:, :, 0], img_lum[:, :, 1], img_lum[:, :, 2]

# Menggunakan bobot tertentu untuk menghitung citra grayscale dari saluran warna
gray_img_lum = (0.24 * R) + (0.77 * G) + (0.06 * B)

# Menampilkan citra asli dan citra grayscale yang dihasilkan
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img_rgb)
ax[1].imshow(gray_img_lum, cmap='gray')
```

5. Tampilkanlah warna tertentu pada citra, dan ubah warna lain menjadi grayscale. Misal,
tampilkan warna biru pada citra masukan dan ubah bagian lain yang tidak berwarna
biru menjadi grayscale seperti pada contoh berikut:

```python
img = cv.imread('/content/drive/MyDrive/lion.JPG')

# Menyaring saluran warna merah dalam gambar dan mengubahnya menjadi gambar biner
ret, mask = cv.threshold(img [:,:,2], 180, 255, cv.THRESH_BINARY)

# Membuat masker tiga saluran dengan saluran merah
mask3 = np.zeros_like(img)
mask3[:,:,0] = mask
mask3[:,:,1] = mask
mask3[:,:,2] = mask

# Menggabungkan saluran merah pada gambar asli
red = cv.bitwise_and(img_rgb, mask3)

# Mengubah gambar asli menjadi skala abu-abu
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img = cv.cvtColor(gray, cv.COLOR_GRAY2BGR)

# Menyaring saluran bukan merah dalam gambar dan menggabungkannya dengan saluran merah
gray = cv.bitwise_and(img, 255 - mask3)
out = gray + red

# Menampilkan gambar asli dan hasilnya menggunakan matplotlib
fig, ax = plt.subplots(1,2)
ax[0].imshow(img_rgb)
ax[1].imshow(out, cmap='gray')
```
